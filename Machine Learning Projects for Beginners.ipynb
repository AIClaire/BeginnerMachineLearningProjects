# Machine Learning Projects for Beginners

In this repo you will find links to different machine learning projects that you can try out as a beginner.

Use the steps identified in the Machine Learning Step by Step Guide to help you solve these different problems.

These projects have been chosen because of the range of skills they show and the datasets are a manageable size. The dataset size means that they are easy to practise on.

If you have a good project that you think should be added here please do so.


## Project 1: Iris Classification

This project is one of the best entry level projects to help you get started with machine learning projects. In the project you are tasked with helping identify different types of Iris flowers from their petal dimensions

### Link to the data: 

[Iris](https://archive.ics.uci.edu/ml/datasets/iris) - Download from UCI Machine Learning Repository

## Project 2: Rice Leaf Diseases

This project contains a small data set of images of different diseases as they show on rice leaves. Your task is to build a classifier to identify them.

### Link to the data: 

[Rice Leaf Diseases](https://archive.ics.uci.edu/ml/datasets/Rice+Leaf+Diseases) - Download from UCI Machine Learning Repository

## Project 3: Video Game Sales Prediction

This project contains data for different video games that you will use to predict the sales of the video game. Pay close attention to the different columns of data you are given and make sure you know which should be inputs and which should be outputs of the algorithm!


### Link to the data: 

[Video Games Data](https://www.kaggle.com/gregorut/videogamesales) - Download from Kaggle

## Project 4: Film Review  Analysis

This project will help you to classify different reviews of films and see how people feel about them.

### Link to the data: 

[Film Reviews](https://archive.ics.uci.edu/ml/datasets/Sentiment+Labelled+Sentences) - Download from UCI Machine Learning Repository

## Project 5: Customer Data Analysis

Here I am giving you two options but your goal is the same. In this project you need to segment your customer data so that you can make predictions about purchases and target marketing campaigns.

### Link to the data: 

[e-commerce dataset](https://www.kaggle.com/carrie1/ecommerce-data) - Download from Kaggle
[Supermarket Customer Data](https://github.com/sowmyacr/kmeans_cluster/blob/master/CLV.csv) - download from GitHub


## Project 6: Stock Price Predictions

In this project you will be able to choose a company from the various ones listed on the Indian Stock exchange and access their data.

Once you have the data you will use it to predict the stock price. This is a time series forecasting project.

### Link to the data: 

[Stock Price Data](https://www.quandl.com/data/NSE-National-Stock-Exchange-of-India) - please note you will need to create an account to access this data.

In the above projects you will have the opportunity to showcase multiple different skills. Remember you can try different algorithms to see which gives you the best result.

Happy experimenting!
